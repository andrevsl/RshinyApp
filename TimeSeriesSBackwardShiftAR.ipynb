```python
set.seed(2016);N=1000;phi=1
Z=rnorm(N,0,1);X=NULL;
X[1]=Z[1]
for (t in 2:N){
  X[t]=Z[t]+phi*X[t-1]
}
```

```python
X.ts=ts(X)
par(mfrow=c(2,1))
plot(X.ts,main="AR(1) Time Series on white Noise, phi=0.4")
X.act=(acf(X.ts,main="AR(1) Time Series on Whte Noise,phi=0.4"))
```

```python
phi1=0.1
Xarima.ts<-arima.sim(list(ar=c(phi1)),n=1000)
par(mfrow=c(2,1))
plot(Xarima.ts,main=paste("AR(1) Time Series, phi1=",phi1))
Xarima.acf=acf(X.ts,main="Autocorrelation of AR(1) Time Series")
```
